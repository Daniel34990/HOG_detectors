```python
from skimage import io
import matplotlib.pyplot as plt
```

```python
import json

# Chemin vers le fichier JSON
file_path = "/Users/daniel/Desktop/Cours Ingé/Cours 2A/IMA/IMA204/HOG_detectors/INRIA_dataset/train/_annotations.coco.json"

# Charger le fichier JSON
with open(file_path, 'r') as f:
    coco_data = json.load(f)

# Afficher les clés principales
print(coco_data.keys())
```

```python
image_path = "/Users/daniel/Desktop/Cours Ingé/Cours 2A/IMA/IMA204/HOG_detectors/INRIA Person detection dataset.v1i.coco/test/crop_000001_jpg.rf.5d8c095a99c1832139d343d79dcf2918.jpg"
```

```python
image_dict = {image['id']: image for image in coco_data['images']}
print(image_dict)
```

```python
def read_image(id):
    image_folder = "/Users/daniel/Desktop/Cours Ingé/Cours 2A/IMA/IMA204/HOG_detectors/INRIA_dataset/train"
    image_info = image_dict[id]
    image_path = f"{image_folder}/{image_info["file_name"]}"
    image = io.imread(image_path)
    return image
```

```python
image = read_image(0)
plt.imshow(image)
plt.show()
```
